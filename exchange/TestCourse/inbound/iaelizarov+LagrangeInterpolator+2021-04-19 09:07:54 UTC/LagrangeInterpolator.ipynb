# Интерполяция при помощи полиномов Лагранжа.

```python
import numpy as np

class LagrangeInterpolator:
    
    def __init__(self, xk, yk):
        self.xk = np.asarray(xk, dtype=float)
        self.yk = np.asarray(yk, dtype=float)
        
    def __call__(self, x):
        Pol = 0
        for i in range(len(self.yk)):
            l_k = 1
            for j in range(len(self.xk)):
                if j != i:
                    l_k *= (x - self.xk[j]) / (self.xk[i] - self.xk[j])
            Pol += l_k * self.yk[i]
        return Pol
```

```python
def runge_func(x, a=25):
    return 1.0 / (1.0 + a*x**2)

xx = np.linspace(-2, 2, 21)
yy = runge_func(xx)

lagr = LagrangeInterpolator(xx, yy)

from numpy.testing import assert_allclose

assert_allclose(yy,
                [lagr(xval) for xval in xx],
                atol=1e-14)
```

```python
import matplotlib.pyplot as plt
fig1, ax1 = plt.subplots(figsize=(8,7))
xx2 = np.linspace(-2, 2, 200)
list_of_m = [3, 5, 7, 11]
for m in list_of_m:
    xx1 = np.linspace(-2, 2, m)
    yy1 = runge_func(xx1)
    lagr = LagrangeInterpolator(xx1, yy1)
    yy2 = [lagr(xval) for xval in xx2]
    ax1.plot(xx2, yy2, '-', label='m = %s' % m)

ax1.plot(xx2, runge_func(xx2), color='k', label='runge func')
plt.legend(loc='best')
plt.grid()
```

```python
def cheb_nodes(n):
    k = np.arange(n)
    return np.cos((2*k + 1) * np.pi / (2*n+2))

fig2, ax2 = plt.subplots(figsize=(8,7))
list_of_m = [3, 5, 7, 11]
for m in list_of_m:
    xx3 = 2 * cheb_nodes(m)
    yy3 = runge_func(xx3)
    lagr = LagrangeInterpolator(xx3, yy3)
    yy2 = [lagr(xval) for xval in xx2]
    ax2.plot(xx2, yy2, '-', label='m = %s' % m)

ax2.plot(xx2, runge_func(xx2), color='k', label='runge func')
plt.legend(loc='best')
plt.grid()
```
