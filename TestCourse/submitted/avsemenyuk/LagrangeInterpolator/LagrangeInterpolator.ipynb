# Интерполяция при помощи полиномов Лагранжа.

```python
import numpy as np

class LagrangeInterpolator:
    """Lagrange interpolating polynomial.
    """

    def __init__(self, xk, yk):
        self.xk = np.asarray(xk, dtype=float)
        self.yk = np.asarray(yk, dtype=float)

    def __call__(self, x):
        """Evaluate the interpolator at a given point.
        """
        n = self.xk.size
        value = 0.0

        for j in range(n):
            frac = 1.0
            for m in range(n):
                if m == j:
                    continue
                num = x - self.xk[m]
                den = self.xk[j] - self.xk[m]
                frac *= num / den
            value += frac * self.yk[j]
        return value
```

```python
def runge_func(x, a=25):
    return 1.0 / (1.0 + a*x**2)

xx = np.linspace(-2, 2, 21)
yy = runge_func(xx)

lagr = LagrangeInterpolator(xx, yy)

from numpy.testing import assert_allclose

assert_allclose(yy,
                [lagr(xval) for xval in xx],
                atol=1e-14)
```

```python
import matplotlib.pyplot as plt
t = np.linspace(-2, 2, 100)
fig, ax = plt.subplots(2,2,figsize=(12,8))
fig.suptitle('Equidistant nodes', fontsize=16, y=0.95)

for i,m in enumerate([3, 5, 7, 11]):
    ax[(i)//2, (i)%2].plot(t, runge_func(t), 'g', label='given function')
    nodes = np.linspace(-2, 2, m)
    lagr = LagrangeInterpolator(nodes, runge_func(nodes))
    ax[(i)//2, (i)%2].plot(nodes, runge_func(nodes), 'ro')
    ax[(i)//2, (i)%2].plot(t, lagr(t), '--m', label='interpolation')
    ax[(i)//2, (i)%2].legend()
    ax[(i)//2, (i)%2].set_title(f'nodes number = {m}')
```

```python
def chebyshev_nodes(a, b, n):
    return [1/2*(a+b) + 1/2*(b-a)*np.cos(np.pi * (2*k - 1) / (2*n)) for k in range(1, n+1)]
```

```python
t = np.linspace(-2, 2, 100)
fig, ax = plt.subplots(2,2,figsize=(12,8))
fig.suptitle('Сhebyshev nodes', fontsize=16, y=0.95)

for i,m in enumerate([3, 5, 7, 11]):
    ax[(i)//2, (i)%2].plot(t, runge_func(t), 'g', label='given function')
    nodes = np.array(chebyshev_nodes(-2, 2, m))
    lagr = LagrangeInterpolator(nodes, runge_func(nodes))
    ax[(i)//2, (i)%2].plot(nodes, runge_func(nodes), 'ro')
    ax[(i)//2, (i)%2].plot(t, lagr(t), '--m', label='interpolation')
    ax[(i)//2, (i)%2].legend()
    ax[(i)//2, (i)%2].set_title(f'nodes number = {m}')
```
