```python
def householder(vec):
    """Ð¡Ð¾Ð·Ð´Ð°ÐµÑ‚ Ð¾Ñ‚Ñ€Ð°Ð¶ÐµÐ½Ð¸Ðµ Ð¥Ð°ÑƒÑÑ…Ð¾Ð»Ð´ÐµÑ€Ð°, Ð¿Ñ€ÐµÐ¾Ð±Ñ€Ð°Ð·ÑƒÑŽÑ‰ÐµÐµ 2-ÑŽ, 3-ÑŽ Ð¸ Ñ‚.Ð´. ÐºÐ¾Ð¼Ð¿Ð¾Ð½ÐµÐ½Ñ‚Ñ‹ Ð²ÐµÐºÑ‚Ð¾Ñ€Ð° vec Ð² Ð½ÑƒÐ»Ð¸.
    
    Parameters
    ----------
    vec : array-like of floats, shape (n,)
        Ð’Ð²ÐµÐ´ÐµÐ½Ð½Ñ‹Ð¹ Ð²ÐµÐºÑ‚Ð¾Ñ€
    
    Returns
    -------
    outvec : array of floats, shape (n,)
        ÐŸÑ€ÐµÐ¾Ð±Ñ€Ð°Ð·Ð¾Ð²Ð°Ð½Ð½Ñ‹Ð¹ Ð²ÐµÐºÑ‚Ð¾Ñ€, Ð¿Ñ€Ð¸Ñ‡ÐµÐ¼ ``outvec[1:]==0`` Ð¸ ``|outvec| == |vec|``
    H : array of floats, shape (n, n)
        ÐžÑ€Ñ‚Ð¾Ð³Ð¾Ð½Ð°Ð»ÑŒÐ½Ð°Ñ Ð¼Ð°Ñ‚Ñ€Ð¸Ñ†Ð° Ð¾Ñ‚Ñ€Ð°Ð¶ÐµÐ½Ð¸Ð¹ Ð¥Ð°ÑƒÑÑ…Ð¾Ð»Ð´ÐµÑ€Ð°
    """
    vec = np.asarray(vec, dtype=float)
    if vec.ndim != 1:
        raise ValueError("vec.ndim = %s, expected 1" % vec.ndim)
        
    u = np.empty_like(vec, dtype=float)
    if vec[0] > 0:
        u[0] = (-1*(vec[1:]**2).sum()) / (vec[0] + np.linalg.norm(vec))
    else:
        u[0] = vec[0] - np.linalg.norm(vec)
    u[1:] = vec[1:]
    u /= np.linalg.norm(u)
    
    H = np.eye(vec.size) - (2 * u.reshape(-1,1) @ u.reshape(1,-1))
    
    return H@vec, H
```

```python
def qr_decomp(a):
    """Ð’Ñ‹Ñ‡Ð¸ÑÐ»ÑÐµÑ‚ QR - Ñ€Ð°Ð·Ð»Ð¾Ð¶ÐµÐ½Ð¸Ðµ Ð¼Ð°Ñ‚Ñ€Ð¸Ñ†Ñ‹.
    
    Parameters
    ----------
    a : ndarray, shape(m, n)
        Ð’Ð²ÐµÐ´ÐµÐ½Ð½Ð°Ñ Ð¼Ð°Ñ‚Ñ€Ð¸Ñ†Ð°
    
    Returns
    -------
    q : ndarray, shape(m, m)
        ÐžÑ€Ñ‚Ð¾Ð³Ð¾Ð½Ð°Ð»ÑŒÐ½Ð°Ñ Ð¼Ð°Ñ‚Ñ€Ð¸Ñ†Ð°
    r : ndarray, shape(m, n)
        Ð’ÐµÑ€Ñ…Ð½ÑÑ Ñ‚Ñ€ÐµÑƒÐ³Ð¾Ð»ÑŒÐ½Ð°Ñ Ð¼Ð°Ñ‚Ñ€Ð¸Ñ†Ð°
        
    Examples
    --------
    >>> a = np.random.random(size=(3, 5))
    >>> q, r = qr_decomp(a)
    >>> np.testing.assert_allclose(q @ r, a, atol=1e-14)
    
    """
    m, n = a.shape
    
    r = np.array(a, copy=True, dtype=float)
    q = np.eye(m)
    
    
    for i in range(n):
        _, H_i = householder(r[i:,i])
        H = np.eye(m)
        H[i:,i:] = H_i
        
        r = H @ r
        q = q @ H
        
    return q, r
```

Ð£Ð±ÐµÐ´Ð¸Ñ‚ÐµÑÑŒ, Ñ‡Ñ‚Ð¾ Ð²Ñ‹ Ð´Ð¾Ð±Ð°Ð²Ð¸Ð»Ð¸ Ð´Ð¾ÑÑ‚Ð°Ñ‚Ð¾Ñ‡Ð½Ð¾ ÐºÐ¾Ð¼Ð¼ÐµÐ½Ñ‚Ð°Ñ€Ð¸ÐµÐ², ÑÐ»ÐµÐ´ÑƒÑŽÑ‰Ð¸Ñ… Ð²Ð°ÑˆÐ¸Ð¼ Ð²Ñ‹ÐºÐ»Ð°Ð´ÐºÐ°Ð¼. 



```python
def qr_nomatrix(a):
    """Form QR decomposition of `a` via successive Householder reflections.
    
    Parameters
    ----------
    a : ndarray
        Input matrix
    
    Returns
    -------
    R : ndarray
        Upper triangular matrix of the QR decomposition
    
    U : ndarray
        Columns store successive Householder reflectors: `U[j:, j]` stores
        the Householder reflector for reducing the `j`-th column.
        
    See Also
    --------
    householder_apply : apply Householder reflectors stored in `U` to `a`.
    
    """
    m, n = a.shape
    R = np.array(a)
    U = np.empty_like(a)
    
    for i in range(n):
        vec = R[i:,i]
        
        u = np.empty_like(vec, dtype=float)
        if vec[0] > 0:
            u[0] = (-1*(vec[1:]**2).sum()) / (vec[0] + np.linalg.norm(vec))
        else:
            u[0] = vec[0] - np.linalg.norm(vec)
        u[1:] = vec[1:]
        u /= np.linalg.norm(u)
        
        U[:,i] = np.append(np.zeros(i), u)
        
        R[i:,i:] -= 2*u.reshape(-1,1) @ (u.reshape(1,-1) @ R[i:,i:])
        
    return R, U

    
def householder_apply(b, uu):
    """Apply the Householder reflectors stored in `uu` to `b`.
    
    The result is equivalent to
    >>> r, q = qr_decomp(a)
    >>> q.T @ b
    
    Parameters
    ----------
    b : ndarray
        Input matrix
    uu : ndarray
        Householder reflectors: `U[j:, j]` is the reflection vector
        for transforming the `j`-th column of `a`.
        
    Returns
    -------
    r : ndarray
        The result of applying the reflectors to `b`. Equivalent to
        computing `q.T @ b`.

    See Also
    --------
    qr_decomp
    
    """
    r = np.asarray(b)
    n = uu.shape[1]
    
    for i in range(n):
        u = uu[i:,i]
        r[i:,i:] -= 2*u.reshape(-1,1) @ (u.reshape(1,-1) @ r[i:,i:])
    
    return r
```

```python
rndm = np.random.RandomState(1223)

a = rndm.uniform(size=(5, 5))
R1, U = qr_nomatrix(a)
R2 = householder_apply(a, U)
R_lib = qr(a)[1]
assert_allclose(R1, R2, atol=1e-14)

###BEGIN HIDDEN TESTS

name2 = "Test 2: vertical rectangular matrix"
a2 = rndm.uniform(size=(5, 3))
R1, U = qr_nomatrix(a2)
R22 = householder_apply(a2, U)
R_lib2 = qr(a2)[1]
assert_allclose(R1, R22, atol=1e-14)


name3 = "Test 3: horizontal rectangular matrix"
a3 = rndm.uniform(size=(3, 5))
R1, U = qr_nomatrix(a3)
R23 = householder_apply(a3, U)
R_lib3 = qr(a3)[1]
assert_allclose(R1, R23, atol=1e-14)

###END HIDDEN TESTS
```
