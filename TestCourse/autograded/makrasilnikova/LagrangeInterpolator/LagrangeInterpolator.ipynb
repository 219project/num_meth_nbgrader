# Интерполяция при помощи полиномов Лагранжа.

```python
import numpy as np

class LagrangeInterpolator:
    """Lagrange interpolating polynomial.
    
    Given a set of pairs ``(x_k, y_k)``, construct 
    a Lagrange polynomial ``f(x)``, such that
    
    .. math::

        f(x_k) = y_k   for k =0, ..., n-1
    
    Parameters
    ----------
    xk : array_like, shape(n,)
        Abscissas
    yk : array_like, shape(n,)
        Ordinates
    
    Attributes
    ----------
    __call__
    
    """
    def __init__(self, xk, yk):
        self.xk = np.asarray(xk, dtype=float)
        self.yk = np.asarray(yk, dtype=float)
        
    def __call__(self, x):
        """Evaluate the interpolator at a given point.
        
        Parameters
        ----------
        x : float
        
        Returns
        -------
        the value of the interpolator at ``x``.
        """
        z=0
        for j in range(len(self.yk)):
            p1=1
            p2=1
            for i in range(len(self.xk)):
                if i!=j:
                  p1=p1*(x-self.xk[i])
                  p2=p2*(self.xk[j]-self.xk[i])
            z=z+self.yk[j]*p1/p2
        return z
```

```python
def runge_func(x, a=25):
    return 1.0 / (1.0 + a*x**2)

xx = np.linspace(-2, 2, 21)
yy = runge_func(xx)

lagr = LagrangeInterpolator(xx, yy)

from numpy.testing import assert_allclose

assert_allclose(yy,
                [lagr(xval) for xval in xx],
                atol=1e-14)
```

```python
from math import pi, cos
```

```python
def Chebyshev_nodes(n, a, b):
  nodes = []
  for i in range(1,n + 1):
    nodes.append(0.5*(a + b) + 0.5*(b - a)*cos(((2*i-1)/(2*n))*pi))
  nodes = np.array(nodes)
  return nodes
```

```python
m = [4,6,8,12]
for i in m:
  x = np.linspace(-2, 2, 100)
  rnge = [runge_func(i) for i in x]
  xx_ = Chebyshev_nodes(i, -2, 2)
  yy_ = runge_func(xx_)
  lagr = LagrangeInterpolator(xx_, yy_)
  y = [lagr(xval) for xval in x]
  plt.subplots()
  plt.plot(x, y, 'b--', label = 'Chebyshev Nodes')
  plt.plot(x, rnge,'m--', label = 'Runge Function')
  plt.legend()
```
